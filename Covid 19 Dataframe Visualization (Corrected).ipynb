```python
# dropping unnessesary columns form the first dataset. if you want to override the dataframe then inplace = True
# axis = 0 represent rows and axis = 1 represent column
covid_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace = True, axis = 1)
```

```python
covid_df
```

```python
# change the fomrmat of the date column

covid_df["Date"] = pd.to_datetime(covid_df["Date"], format = '%Y-%m-%d')
```

```python
# to find total number of active cases, we have to subtract
#from total number of confirmed cases to sum of (cured cases + death reports)
```

```python
# Active Cases

covid_df['active_cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
```

```python
covid_df.tail()
```

```python
# making a pivot table using pandas library
# index() method helps you find the index position of an element or an item in a string of characters or a list of items.
# aggfunc is column to aggregate and value is function or list of functions.

statewise = pd.pivot_table(covid_df, values = ["Confirmed", "Deaths", "Cured"], index = "State/UnionTerritory", aggfunc = max)
```

```python
# making a recovery rate in percentage
# so we have to multiply the (cured/ confirmed) and multipled with 100


statewise["Recovery Rate"] = statewise["Cured"]*100/statewise["Confirmed"]
```

```python
# making a mortality rate in percentage
# so we have to multiply the (deaths/confirmed) * 100


statewise["Mortality Rate"] = statewise["Deaths"]*100/statewise["Confirmed"]
```

```python
# sort the values in the basis of confirmed columns in descending order

statewise = statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
# plotting a pivot table using a nice visual
# for this we have to use background_gradient function
# cmap stands for colour map which present in the matplotlib library

statewise.style.background_gradient(cmap = "cubehelix")
```

```python
# top 10 active cases states
# for this query we have to use another pandas function which is called as groupby
# we group by the columns by states followed by max() which can help us to find maximum values in states that have active cases and 
#we sort the values by active cases in descending order and then reset the index 
# for that we have to use the reset_index() function


Top_10_Active_Cases = covid_df.groupby(by = "State/UnionTerritory").max()[["active_cases", "Date"]].sort_values(by = ["active_cases"], ascending = False).reset_index() 
```

```python
Top_10_Active_Cases
```

```python
Top_10_Active_Cases = covid_df.groupby(by = "State/UnionTerritory").max()[["active_cases", "Date"]].sort_values(by = ["active_cases"], ascending = False).reset_index() 
fig = plt.figure(figsize = (16,9))
plt.title("Top 10 States", size = 25)
ax = sns.barplot(data = Top_10_Active_Cases.iloc[:10], y = "active_cases", x = "State/UnionTerritory", linewidth = 2, edgecolor = "red")
plt.xlabel("States")
plt.ylabel("Total Active Cases")
plt.show()
```

```python
# now we will se the top 10 states on the basis of deaths columns

top_10_deaths = covid_df.groupby(by = "State/UnionTerritory").max()[["Deaths", "Date"]].sort_values(by = ["Deaths"], ascending = False).reset_index()
```

```python
fig = plt.figure(figsize = (18,5))
plt.title("Top 10 States for Deaths", size = 25)
ax = sns.barplot(data = top_10_deaths.iloc[:12], y = "Deaths", x = "State/UnionTerritory", linewidth = 2, edgecolor = "black")
plt.xlabel("States")
plt.ylabel("Total Death Cases")
plt.show()
```

```python
# remove states where states is India

vaccine = vaccine_df[vaccine_df.State != 'India']
```

```python
# Most Vaccinated State
# for this querry we need to use groupby function

max_vac = vaccine.groupby('State')['Total'].sum().to_frame('Total')
max_vac = max_vac.sort_values('Total', ascending = False)[:5]
max_vac
```

```python
fig = plt.figure(figsize = (10,5))
plt.title('Top 5 Vaccinated State in India', size = 20)
x = sns.barplot(data = max_vac.iloc[:10], y = max_vac.Total, x = max_vac.index, linewidth = 2, edgecolor = 'black')
plt.xlabel = ('States')
plt.ylabel = ('Vaccination')
plt.show
```
